```python
with open(meta_path,'rb') as file:
    data = pk.load(file,encoding='latin1')
    class_names = data['fine_label_names']
print(class_names)
```

```python
def plot_images(images,labels,class_name):
    plt.figure(figsize=(12,12))
    for i in range(25):
        plt.subplot(10,5,i+1)
        plt.imshow(images[i])
        plt.title(class_name[labels[i][0]])
        plt.axis('off')
    plt.show()
```

# check the distribution of the classes

```python
def visualize_distribution(data,name):
    labels_df = pd.DataFrame(data,columns=['label'])
    label_count = labels_df['label'].value_counts().sort_index()

    # plot and visualize the class distribution

    plt.figure(figsize=(10,10))
    sns.barplot(x=label_count.index,y=label_count.values)
    plt.title(f'Distribution of {name} class in CIFAR100 dataset')
    plt.xlabel('Class')
    plt.ylabel('Number of Images')
    plt.tight_layout()
    plt.show()
```

```python
visualize_distribution(y_train,'Train Dataset')
```

```python
visualize_distribution(y_test,'Test Dataset')
```
